```python
import matplotlib.pyplot as plt
import statistics

#série numérique X
consommation_graisse_en_kg_par_an = [14.4, 16, 11.6]
#série numérique Y
taux_de_mortalité_par_maladie_pour_100mille_habitants = [29.1, 29.7, 29.2]

plt.plot(consommation_graisse_en_kg_par_an,taux_de_mortalité_par_maladie_pour_100mille_habitants,"ob") # ob = type de points "o" ronds, "b" bleus
plt.ylabel('Consommation de graisse en kg par an et par personne ')
plt.xlabel('Taux de mortalité par maladie pour 100 000 habitants')
plt.show()

#Construire nuage de point
```

```python
# Calculer les caractéristiques numériques des variables X et Y (moyenne,variance et écart-type).

#moyenne
x_mean = statistics.mean(consommation_graisse_en_kg_par_an)
y_mean = statistics.mean(taux_de_mortalité_par_maladie_pour_100mille_habitants)

#variance
x_variance = statistics.variance(consommation_graisse_en_kg_par_an)
y_variance = statistics.variance(taux_de_mortalité_par_maladie_pour_100mille_habitants)

#écart type
x_standard_deviation = statistics.stdev(consommation_graisse_en_kg_par_an)
y_standard_deviation = statistics.stdev(taux_de_mortalité_par_maladie_pour_100mille_habitants)
```

```python
#Calculer la covariance entre X et Y.
x_y_covariance = statistics.covariance(consommation_graisse_en_kg_par_an, taux_de_mortalité_par_maladie_pour_100mille_habitants)
x_y_correlation = (x_y_covariance)/((x_standard_deviation)*(y_standard_deviation))
```

```python
x_y_correlation
#0.67 : coefficient de correlation au dessus de la moyenne il existe une correlation entre la consommation de gras et l'apparition de la maladie de l'athérosclérose
```

```python
#Calculez le coefficient de corrélation linéaire et de détermination et interprétez les.
a_model_regression = x_y_covariance/x_variance
b_model_regression = y_mean - (x_mean*a_model_regression)
```

```python
#calculer le y ajusté
import numpy as np

y_adjusted = np.array(consommation_graisse_en_kg_par_an) * a_model_regression # Convert the list to a NumPy array for element-wise multiplication

y_adjusted

y_adjusted = y_adjusted + b_model_regression
```

```python
#construire le modele de regression avec le nuage de point
plt.plot(consommation_graisse_en_kg_par_an,taux_de_mortalité_par_maladie_pour_100mille_habitants,"ob") # ob = type de points "o" ronds, "b" bleus
plt.ylabel('Consommation de graisse en kg par an et par personne ')
plt.xlabel('Taux de mortalité par maladie pour 100 000 habitants')
plt.plot(consommation_graisse_en_kg_par_an, y_adjusted)
plt.show()
```

```python
coef_determiation_part_1 = 0
coef_determiation_part_2 = 0
val_tampon = 0
for k in range(3):
    val_tampon = 0
    val_tampon = y_adjusted[k]
    val_tampon -= y_mean
    val_tampon *= val_tampon
    coef_determiation_part_1 += val_tampon
for k in range(3):
    val_tampon = 0
    val_tampon = taux_de_mortalité_par_maladie_pour_100mille_habitants[k]
    val_tampon -= y_mean
    val_tampon *= val_tampon
    coef_determiation_part_2 += val_tampon

coef_determiation = coef_determiation_part_1/coef_determiation_part_2
print(coef_determiation)

#coef deetermination = 0.45 (45%) ce modele de regression n'est pas fiable pour faire des estimation
```
