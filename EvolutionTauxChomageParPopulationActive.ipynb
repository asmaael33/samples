```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics

#série numérique X
active_population_rate = [40,45,50]
#série numérique Y
chomage_rate = [2.1, 3.1, 3.4]

df = pd.DataFrame({'active_population_rate':active_population_rate, 'chomage_rate':chomage_rate})
df
```

```python
# Calculer les caractéristiques numériques des variables X et Y (moyenne,variance et écart-type).

#moyenne
x_mean = statistics.mean(active_population_rate)
y_mean = statistics.mean(chomage_rate)

#variance
x_variance = statistics.variance(active_population_rate)
y_variance = statistics.variance(chomage_rate)

#écart type
x_standard_deviation = statistics.stdev(active_population_rate)
y_standard_deviation = statistics.stdev(chomage_rate)
```

```python
# Calculer a,b,r
#Calculer la covariance entre X et Y.
x_y_covariance = statistics.covariance(active_population_rate, chomage_rate)
x_y_correlation = (x_y_covariance)/((x_standard_deviation)*(y_standard_deviation))

#Calculez le coefficient de corrélation linéaire et de détermination et interprétez les.
a_model_regression = x_y_covariance/x_variance
b_model_regression = y_mean - (x_mean*a_model_regression)
```

```python
x_y_correlation
```

```python
#calculer le y ajusté
import numpy as np

y_adjusted = np.array(active_population_rate) * a_model_regression # Convert the list to a NumPy array for element-wise multiplication

y_adjusted

y_adjusted = y_adjusted + b_model_regression
```

```python
#construire le modele de regression avec le nuage de point
plt.plot(active_population_rate,chomage_rate,"ob") # ob = type de points "o" ronds, "b" bleus
plt.ylabel('évolution du taux de chômage')
plt.xlabel('pourcentage de la population active, au japon, entre 1990 et 2000')
plt.plot(active_population_rate, y_adjusted)
plt.show()
```

```python
coef_determiation_part_1 = 0
coef_determiation_part_2 = 0
val_tampon = 0
for k in range(3):
    val_tampon = 0
    val_tampon = y_adjusted[k]
    val_tampon -= y_mean
    val_tampon *= val_tampon
    coef_determiation_part_1 += val_tampon
for k in range(3):
    val_tampon = 0
    val_tampon = chomage_rate[k]
    val_tampon -= y_mean
    val_tampon *= val_tampon
    coef_determiation_part_2 += val_tampon

coef_determiation = coef_determiation_part_1/coef_determiation_part_2
print(coef_determiation)
```

```python
#Pour la série numérique des taux de chômage trouver les valeurs suivants:médiane, 1er quartiles, 3ème quartiles, Valeur max, Valeur min
#mediane
med_active_population_rate = statistics.median(active_population_rate)
med_active_population_rate
#1er quartille
first_quarter_active_population_rate = np.percentile(active_population_rate, 25)
first_quarter_active_population_rate
#3rd quartille
third_quarter_active_population_rate = np.percentile(active_population_rate, 75)
third_quarter_active_population_rate
```
